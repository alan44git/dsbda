```python
import pandas as pd
```

```python
df = pd.read_csv("../../datasets/Movie/movies_metadata.csv")
df.columns
```

```python
df["original_language"].unique()
```

```python
english = df[df["original_language"]=='en']
french = df[df["original_language"]=='fr']
```

```python
merged = pd.concat([english,french])
```

```python
sorte = merged.sort_values(by='vote_average',ascending=False)
```

```python
transposed = sorte.T
```

```python
melted = pd.melt(
    df,
    id_vars=['genres'],
    value_vars=['revenue', 'runtime']
)
```

```python
wide = melted.pivot_table(
    index=['genres'],
    columns='variable',
    values='value',
)
```
