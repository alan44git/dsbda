```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
columns = ['class','age','sex','steroid','antiviral','fatigue','malaise','anorexia','liver_big','liver_firm','spleen_palpable','spiders','ascites','varices','bilirubin','alk_phospate','sgot','albumin','protime','histology']
df = pd.read_csv("../../datasets/Hepatitis/hepatitis.csv", header=None, names=columns, index_col=None)
```

```python
df.replace('?',np.nan,inplace=True)
df = df.apply(pd.to_numeric,errors='coerce')
```

```python
df.dtypes
```

```python
cols=['steroid','fatigue','malaise','anorexia','liver_big','liver_firm','spleen_palpable','spiders','ascites','varices','bilirubin','alk_phospate','sgot','albumin','protime']
for col in cols:
    df[col]=df[col].fillna(df[col].median())
```

```python
df.isnull().sum()
```

```python
for col in df.columns:
    if(col=='class'): continue
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bund = Q1 - 1.5*IQR
    upper_bund = Q3 + 1.5*IQR
    df[col] = np.where(
        df[col] < lower_bund,
        lower_bund,
        np.where(
            df[col] > upper_bund,
            upper_bund,
            df[col]
        )
    )
```

```python
X = df.drop(columns=['class'])
Y = df['class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,Y,
    random_state=0,
    test_size=0.2,
)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
lr_pred = lr.predict(X_test)
```

```python
nb = GaussianNB()
nb.fit(X_train, y_train)
nb_pred = nb.predict(X_test)
```

```python
print(classification_report(y_test, lr_pred))
print(classification_report(y_test, nb_pred))
```
