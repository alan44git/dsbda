```python
Links = df[df["Type"]=="Link"]
Videos = df[df["Type"]=="Video"]
```

```python
merged = pd.concat([Links, Videos])
```

```python
melted = pd.melt(merged,id_vars=['Type', 'Category'], value_vars=['like','share','comment'])
```

```python
melted.head(100)
```

```python
wide = melted.pivot_table(
    index=['Type','Category'],
    columns='variable',
    values='value',
    aggfunc='sum'
)
```
