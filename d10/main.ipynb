```python
import pandas as pd
```

```python
col = [
    "sepal_length","sepal_width","petal_length","petal_width",
    "class" 
]
df = pd.read_csv("../../datasets/Iris/Iris.csv", header=None,names=col)
```

```python
df.head()
```

```python
setosa = df[df["class"]=="Iris-setosa"]
versicolor = df[df["class"]=="Iris-versicolor"]
```

```python
merged = pd.concat([setosa, versicolor])
```

```python
sorte = merged.sort_values(by='petal_length',ascending=False)
```

```python
transposed = sorte.T
```

```python
melted = pd.melt(
    merged,
    id_vars=['class'],
    value_vars=['sepal_length', 'sepal_width','petal_length','petal_width']
)
```

```python
wide = melted.pivot_table(
    index=['class'],
    columns='variable',
    values='value',
)
```
