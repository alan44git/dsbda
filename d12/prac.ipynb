```python
import pandas as pd
```

```python
df = pd.read_csv("../../datasets/Adult/adult.csv", header=None, names=['age', 'workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','salary'])
df.head()
```

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
```

```python
# a. Create data subsets for different Country, Sex, Race
country_subset = df[df["native-country"]==' United-States']
sex_subset = df[df["sex"]==' Male']
race_subset = df[df["race"]==' White']
```

```python
print("Country Subset (United States):")
print(country_subset.head())
print(f"\nShape: {country_subset.shape}\n")

print("Sex Subset (Male):")
print(sex_subset.head())
print(f"\nShape: {sex_subset.shape}\n")

print("Race Subset (White):")
print(race_subset.head())
print(f"\nShape: {race_subset.shape}")
```

```python
# b. Merge two subsets
merged = pd.concat([sex_subset, race_subset])
```

```python
# c. Sort Data using relevant numeric column (age)
sorted_data = merged.sort_values(by='age', ascending=False)
```

```python
# d. Transposing Data
transposed = sorted_data.T
```

```python
# e. Melting Data to long format
melted = pd.melt(
    merged,
    id_vars=['sex', 'race'],
    value_vars=['age', 'hours-per-week']
)
```

```python
# f. Casting data to wide format (pivot table)
wide = melted.pivot_table(
    index=['sex'],
    columns='variable',
    values='value',
    aggfunc='mean'
)
```
